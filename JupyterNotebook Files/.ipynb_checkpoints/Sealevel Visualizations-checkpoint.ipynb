# Sea Level Data Visualizations
This Takes the original Source data from NASA for one visualization option and takes the other scrubbed data to give a visualization of the complete timeline of sea data. This is because Sea level data can be easily misrepresented. Outliers may be caused from the variety of satellites used over the past and seasonal sea patterns. NASA Scientists state ice loss is the largest contributor to sea level rise but ice melting data is in ShapeFile format. Analyzing fewer years gave a completely opposite picture to sea level data.

```python
# Necessary Imports
%matplotlib inline 

import numpy as np
import requests 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy.linalg as lin
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from scipy import stats
from scipy.stats import zscore
from io import StringIO
```

```python
# Read Dataset
sea_level_data_all = pd.read_csv("../Data/sealevel_data__all_1993-2020.csv") 
sea_level_data_all
```

```python
# Convert Values Into Array
X = sea_level_data_all.iloc[:, 1].values.reshape(-1, 1)

# Calculate Dimension Of Rows
Y = sea_level_data_all.iloc[:, 0].values.reshape(-1, 1)

# Create Object For Class
linear_regressor = LinearRegression()

# Perform Linear Regression
linear_regressor.fit(X, Y)

# Make Predictions
Y_pred = linear_regressor.predict(X)
```

```python
# Plot & Save
plt.scatter(X, Y)
plt.plot(X, Y_pred, color = "red")
plt.xlabel("Years")
plt.ylabel("Global Mean Sea Level (GMSL) in (mm)")
plt.title("Sea Level OverTime (1993-2020) in (mm)")
fig1 = plt.gcf()
fig1.savefig("../Images/Sealevelovertime1993-2020.png", bbox_inches = "tight")
```

This visualization shows the model looking at every datapoint taken for the global mass sea level average as it was measured each day.

```python
# Convert Values Into Array
X = sea_level_data.iloc[:, 0].values.reshape(-1, 1)

# Calculate Dimension Of Rows
Y = sea_level_data.iloc[:, 1].values.reshape(-1, 1)

# Create Object For Class
linear_regressor = LinearRegression()

# Perform Linear Regression
linear_regressor.fit(X, Y)

# Make Predictions
Y_pred = linear_regressor.predict(X)
```

The sea level from 1993 - 2015 showing inconclusive trend over time.

The GMSL is a 1-dimensional time series of globally averaged Sea Surface Height Anomalies (SSHA) from TOPEX/Poseidon, Jason-1, OSTM/Jason-2 and Jason-3. It starts in September 1992 to present, with a lag of up to 4 months. All biases and cross-calibrations have been applied to the data so SSHA are consistent between satellites. Data are reported as changes relative to January 1, 1993 and are 2-month averages. Glacial Isostatic Adjustment (GIA) has been applied. These data are available in ASCII format. Reference: Beckley et al., 2017,

This is the â€œuncertainty margin,â€ or the range from the mean (average) within which there is a high probability that the true number resides. For example, when we say that global mean sea level for a particular month is 50mm +/- 4 mm, the true value for that month has a high probability of being between 46 and 54 mm. The uncertainty margin exists both because of sea level variability (sea level above the first point in the time series) in the ocean and the accuracy with which we measure the individual values. https://sealevel.nasa.gov/understanding-sea-level/key-indicators/global-mean-sea-level/ Source: https://podaac-tools.jpl.nasa.gov/drive/files/allData/merged_alt/L2/TP_J1_OSTM/global_mean_sea_level/GMSL_TPJAOS_4.2_199209_202001.txt

# Removing Outliers

```python
z_scores = stats.zscore(sea_level_data)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 2).all(axis = 1)

sea_level_data
```

```python
sealevel_filtered_df = sea_level_data[filtered_entries]
sealevel_filtered_df
sealevel_filtered_df.to_csv("../Data/RmOutliersGroupedSealevelData.csv", index = False, header = True)
```

```python
# Convert Values Into Array
X = sealevel_filtered_df.iloc[:, 0].values.reshape(-1, 1)

# Calculate Dimension Of Rows
Y = sealevel_filtered_df.iloc[:, 1].values.reshape(-1, 1)

# Create Object For Class
linear_regressor = LinearRegression()

# Perform Linear Regression
linear_regressor.fit(X, Y)

# Make Predictions
Y_pred = linear_regressor.predict(X)
```

```python
# Plot
plt.scatter(X, Y)
plt.plot(X, Y_pred, color = "red")
plt.xlabel("Years")
plt.ylabel("Global Mean Sea Level (GMSL) In (mm)")
plt.title("Sealevel OverTime (in mm) Removed Outliers")
fig3 = plt.gcf()
fig3.savefig("../Images/SealevelRemovedOutliers1993-2015.png", bbox_inches = "tight")
```

```python
sealevel_all_data_df = pd.read_csv("../Data/sealevel_data__all_1993-2020.csv") 

z_scores = stats.zscore(sealevel_all_data_df)
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 2.5).all(axis = 1)

sealevel_all_data_df
```

```python
new_sealevel_df = sealevel_all_data_df[filtered_entries]
new_sealevel_df.to_csv("../Data/RmOutliersSealevel1993-2020.csv", index = False, header = True)
new_sealevel_df
```

```python
# Convert Values Into Array
X = new_sealevel_df.iloc[:, 1].values.reshape(-1, 1)

# Calculate Dimension Of Rows
Y = new_sealevel_df.iloc[:, 0].values.reshape(-1, 1)

# Create Object For Class
linear_regressor = LinearRegression()

# Perform Linear Regression
linear_regressor.fit(X, Y)

# Make Predictions
Y_pred = linear_regressor.predict(X)
```

```python
# Plot
plt.scatter(X, Y)
plt.plot(X, Y_pred, color = "red")
plt.xlabel("Years")
plt.ylabel("Global Mean Sea Level (GMSL) In (mm)")
plt.title("Sealevel OverTime (in mm) Removed Outliers")
fig4 = plt.gcf()
fig4.savefig("../Images/SealevelRemovedOutliers1993-2020.png", bbox_inches = "tight")
```
