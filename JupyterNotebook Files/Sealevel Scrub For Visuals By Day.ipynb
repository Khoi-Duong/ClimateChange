# Sea Level Data Scrub for a second file of data with day by day points
### Source NASA dataset but retaining the datapoints by day and creates a csv for multiple visualizations.
* This notebook takes the origianl source from NASA and creates a file that can be used for other visualizations.

```python
# Necessary Imports
%matplotlib inline 

import requests 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy.linalg as lin
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
from io import StringIO
from datetime import timedelta, datetime
```

```python
# Read Data
sea_data_df = pd.read_csv("../Data/GMSL_merged_nasa_1993_2020.csv")
sea_data_df
```

```python
# Conversion Formula For Date
def convert_partial_year(number):
    year = int(number)
    d = timedelta(days = (number - year) * 365.25)
    day_one = datetime(year, 1, 1)
    date = d + day_one
    return date
```

```python
sea_data_df["date"] = sea_data_df["year_fraction"].apply(convert_partial_year)
sea_data_df = sea_data_df.drop(["year_fraction"], axis = 1)
sea_data_df["year"] = sea_data_df["date"].apply(lambda x : x.year)
sea_data_df
```

```python
sea_data_df = sea_data_df.drop(["date"], axis=1)
sea_data_df.columns = ["GMSL", "Year"]
sea_data_df
```

```python
# Saving Data To CSV
sea_data_df.to_csv("../Data/sealevel_data__all_1993-2020.csv", index = False, header = True)
```
