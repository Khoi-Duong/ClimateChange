# CSV File For Prophet:
### Temperature Population Sea Level and Emissions
### Notebook To Combine CSV Datasets

```python
# Necessary Imports
import pandas as pd
import numpy as np
```

```python
# File to Load  
temperature_data = "../Data/temp_data.csv"
population_data = "../Data/population_data.csv"
sealevel_data = "../Data/sealevel_data.csv"
co2_data = "../Data/co2_ppm.csv"
datestamp_data = "../Data/datestamp.csv"

# Read Data  
temperature_data_df = pd.read_csv(temperature_data)
population_data_df = pd.read_csv(population_data)
sealevel_data_df = pd.read_csv(sealevel_data)
co2_data_df = pd.read_csv(co2_data)
datestamp_data_df = pd.read_csv(datestamp_data)
```

```python
# Merage Files
first_merged_table = pd.merge(temperature_data_df, population_data_df, on = "Year", how = "left")
second_merged_table = pd.merge(first_merged_table, sealevel_data_df, on = "Year", how = "left")
climate_merged_df = pd.merge(second_merged_table, co2_data_df, on = "Year", how = "left")
climate_merged_df = climate_merged_df.drop(["Unnamed: 0"], axis = 1)
climate_merged_df["DS"] = datestamp_data_df
climate_merged_df
```

```python
# Save Combined Dataset To CSV
climate_merged_df.to_csv("../Data/combinedclimatedata_prophet.csv")
```
